```python
df = spark.table("workspace.default.amazon_products")
display(df.limit(5))
print(df.count())

```

```python
df.printSchema()
print("Rows:", df.count())
display(df.limit(5))
```

```python
display(df.select("final_price", "initial_price").limit(10))

```

```python
products = df \
    .withColumn("final_price_clean", F.regexp_replace(F.col("final_price"), '"', '')) \
    .withColumn("initial_price_clean", F.regexp_replace(F.col("initial_price"), '"', '')) \
    .withColumn("final_price_num", F.expr("try_cast(final_price_clean as double)")) \
    .withColumn("initial_price_num", F.expr("try_cast(initial_price_clean as double)")) \
    .withColumn("reviews_count_num", F.expr("try_cast(reviews_count as int)")) \
    .withColumn("rating_num", F.expr("try_cast(rating as double)"))
```

```python
display(products.select("title","final_price","final_price_clean","final_price_num",
                        "initial_price","initial_price_clean","initial_price_num").limit(10))
```

```python
products.select(
    F.count("*").alias("rows"),
    F.sum(F.col("final_price_num").isNull().cast("int")).alias("null_final_price"),
    F.sum(F.col("initial_price_num").isNull().cast("int")).alias("null_initial_price")
).show()
```

```python
## Joins 
```

```python
brand_dim = products.select("brand").dropna().dropDuplicates() \
    .withColumn("brand_id", F.monotonically_increasing_id())

print("Total brands:", brand_dim.count())
display(brand_dim.limit(10))
```

```python
inner_joined = products.join(brand_dim, on="brand", how="inner")
print("Inner join rows:", inner_joined.count())
```

```python
left_joined = products.join(brand_dim, on="brand", how="left")
print("Left join rows:", left_joined.count())
```

```python
outer_joined = products.join(brand_dim, on="brand", how="outer")
print("Outer join rows:", outer_joined.count())
```

```python
display(left_joined.filter(F.col("brand_id").isNull()).select("title","brand").limit(10))
```

```python
## Window Functions 
```

```python
w_brand_price = Window.partitionBy("brand").orderBy(F.desc("final_price_num"))

ranked = products.withColumn("price_rank_in_brand", F.dense_rank().over(w_brand_price))
```

```python
display(
    ranked.filter(F.col("brand").isNotNull())
          .select("brand","title","final_price_num","price_rank_in_brand")
          .orderBy("brand","price_rank_in_brand")
          .limit(30)
)
```

```python
top3_per_brand = ranked.filter(F.col("price_rank_in_brand") <= 3)
display(top3_per_brand.select("brand","title","final_price_num","price_rank_in_brand")
                      .orderBy("brand","price_rank_in_brand"))
```

```python
w_running = Window.partitionBy("brand") \
                  .orderBy("timestamp") \
                  .rowsBetween(Window.unboundedPreceding, Window.currentRow)

running_total = products.withColumn("running_total_price_brand",
                                    F.sum("final_price_num").over(w_running))
display(
    running_total.filter(F.col("brand").isNotNull())
                 .select("brand","timestamp","final_price_num","running_total_price_brand")
                 .orderBy("brand","timestamp")
                 .limit(25)
)
```

```python
## Derived Features 
```

```python
features = products.withColumn(
    "discount_percent",
    F.when(
        (F.col("initial_price_num") > 0) & (F.col("final_price_num").isNotNull()),
        ((F.col("initial_price_num") - F.col("final_price_num")) / F.col("initial_price_num")) * 100
    ).otherwise(None)
)
```

```python
features = features.withColumn(
    "price_bucket",
    F.when(F.col("final_price_num") < 20, "low") \
     .when((F.col("final_price_num") >= 20) & (F.col("final_price_num") < 100), "mid") \
     .otherwise("high")
)
```

```python
features = features.withColumn(
    "review_level",
    F.when(F.col("reviews_count_num") >= 1000, "viral") \
     .when(F.col("reviews_count_num") >= 100, "popular") \
     .otherwise("new")
)
```

```python
display(features.select("title","final_price_num","initial_price_num","discount_percent",
                        "price_bucket","reviews_count_num","review_level").limit(20))
```

```python
## UDF Section
```

```python
from pyspark.sql.types import StringType

def get_main_category(cat):
    if cat is None:
        return None
    return str(cat).split(",")[0].strip()

main_category_udf = F.udf(get_main_category, StringType())
```

```python
features = features.withColumn("main_category", main_category_udf(F.col("categories")))
display(features.select("title","categories","main_category").limit(10))
```

```python
top5 = features.groupBy("asin", "title") \
    .agg(F.sum("final_price_num").alias("revenue")) \
    .orderBy(F.desc("revenue")) \
    .limit(5)

display(top5)
```

```python
pivot_stats = features.groupBy("main_category") \
    .pivot("price_bucket") \
    .count() \
    .orderBy(F.desc("high"))

display(pivot_stats)
```
